```python
import pandas as pd
df = pd.read_csv('Day_10_banking_data.csv')
filtered_by_transaction_amount = df[df['Transaction_Amount'] <= 2000]
print("Rows where Transaction_Amount is <= 2000:")
print(filtered_by_transaction_amount)
loan_payment_high_balance = df[(df['Transaction_Type'] == 'Loan Payment') & (df['Account_Balance'] > 5000)]
print("\nRows with 'Loan Payment' and Account_Balance > 5000:")
print(loan_payment_high_balance)
uptown_branch_transactions = df[df['Branch'] == 'Uptown']
print("\nTransactions made in the 'Uptown' branch:")
print(uptown_branch_transactions)
```

```python
df['Transaction_Fee'] = df['Transaction_Amount'] * 0.02
print("\nData with the new 'Transaction_Fee' column:")
print(df[['Transaction_Amount', 'Transaction_Fee']])s
df['Balance_Status'] = df['Account_Balance'].apply(lambda x: 'High Balance' if x > 5000 else 'Low Balance')
print("\nData with the new 'Balance_Status' column:")
print(df[['Account_Balance', 'Balance_Status']])
```
